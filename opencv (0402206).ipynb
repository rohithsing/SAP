# Read and Display an Image

```python
import cv2
import matplotlib.pyplot as plt

# ---------- Utility function ----------
def show(img, title="Image", gray=False):
    plt.figure(figsize=(4,4))
    if gray:
        plt.imshow(img, cmap="gray")
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()

    
# ---------- Read image ----------
img = cv2.imread(r"C:\Users\rohit\OneDrive\Desktop\learn\SAP\sher.jpg")

if img is None:
    raise ValueError("Image not found or unsupported format")

show(img, "Original Image")

```

### Convert Image to Grayscale

```python

# ---------- Grayscale ----------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(gray, "Grayscale", gray=True)
```

```python
# gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow(gray)
```

## Resize an Image

```python

# ---------- Resize ----------
resized = cv2.resize(img, (500, 300))
show(resized, "Resized Image")
```

```python
# resized = cv2.resize(img, (500, 300))   # width, height
# cv2.imshow(resized)
```

## Edge Detection (Canny)

```python

# ---------- Edge Detection ----------
edges = cv2.Canny(gray, 100, 200)
show(edges, "Canny Edges", gray=True)
```

## Horizontal Flip (Left  Right)

```python

# ---------- Flips ----------
h_flip = cv2.flip(img, 1)
show(h_flip, "Horizontal Flip")
```

```python
# h_flip = cv2.flip(img, 1)
# cv2_imshow(h_flip)
```

## Vertical Flip (Top  Bottom)

```python

v_flip = cv2.flip(img, 0)
show(v_flip, "Vertical Flip")
```

```python
# # v_flip = cv2.flip(img, 0)
# cv2_imshow(v_flip)
```

## Rotate 90 Clockwise

```python

# ---------- Rotation ----------
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
show(rotated, "Rotated 90")
```

```python
# rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
# cv2_imshow(rotated)
```

```python

# ---------- Channel Split ----------
b, g, r = cv2.split(img)
show(b, "Blue Channel", gray=True)
show(g, "Green Channel", gray=True)
show(r, "Red Channel", gray=True)
```

```python
# b, g, r = cv2.split(img)

# cv2_imshow(b)
# cv2_imshow(g)
# cv2_imshow(r)
```

```python

# ---------- Merge Channels ----------
merged = cv2.merge([b, g, r])
show(merged, "Merged Image")
```
